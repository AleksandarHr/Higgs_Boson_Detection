````python
# Median is more robust to outliers
df_train_0_ME1 = df_train_0.DER_mass_MMC.fillna(df_train_0.DER_mass_MMC.median(), inplace=False)
df_train_1_ME1 = df_train_1.DER_mass_MMC.fillna(df_train_1.DER_mass_MMC.median(), inplace=False)
df_train_2_ME1 = df_train_2.DER_mass_MMC.fillna(df_train_2.DER_mass_MMC.median(), inplace=False)
df_train_3_ME1 = df_train_3.DER_mass_MMC.fillna(df_train_3.DER_mass_MMC.median(), inplace=False)
````
